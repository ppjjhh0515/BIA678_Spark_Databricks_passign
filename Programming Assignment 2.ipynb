```
df = sqlContext.read.format("csv").option("inferSchema", "true").option("header", "false").load("dbfs:/FileStore/shared_uploads/jpark28@stevens.edu/Class_9___12____Data_for_Programming___Environmental___v___assignment.csv") 
df = df.drop(*["_c14", "_c16"])
df = df.na.drop(thresh = 12)
df.show()
```

```
#Seperate DF into temperature and precip values
avg_temp_data = df.filter(df['_c0'] == "Average Temperature (F)").drop(df['_c0'])
avg_temp_data.show()
avg_precip_data = df.filter(df['_c0'] == "Average Precipitation (in)").drop(df['_c0'])
avg_precip_data.show()
```

```
#change column names
from functools import reduce

oldColumns = avg_temp_data.schema.names
newColumns = ['ANNUAL', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', '# CITIES']

avg_temp_data = reduce(lambda avg_temp_data, idx: avg_temp_data.withColumnRenamed(oldColumns[idx], newColumns[idx]), range(len(oldColumns)), avg_temp_data)
avg_temp_data.show()

oldColumns = avg_precip_data.schema.names

avg_precip_data = reduce(lambda avg_precip_data, idx: avg_precip_data.withColumnRenamed(oldColumns[idx], newColumns[idx]), range(len(oldColumns)), avg_precip_data)
avg_precip_data.show()
```

```
#String to float
from pyspark.sql.functions import col

for c in avg_temp_data.columns:
  avg_temp_data = avg_temp_data.withColumn(c, col(c).cast('float'))

for c in avg_precip_data.columns:
  avg_precip_data = avg_precip_data.withColumn(c, col(c).cast('float'))

avg_temp_data.dtypes
avg_precip_data.dtypes
```

```
#Function to calculate average annual and monthly temperature/precipitation
from pyspark.sql.functions import round

def get_mean_agg_region(df):
  # Muliply all values with # of cities in each region
  for column in df.schema.names:
    if column != '# CITIES': 
      df = df.withColumn(column, (col(column) * col('# CITIES')))
  
  #check whether data chaged properly
  df.show()    
  
  # Aggregate sum of total values of different region
  df = df.groupBy().sum() 
  
  # Divide aggregated value to total # CITIES
  for column in df.schema.names:
    if column != 'sum(# CITIES)':
      df = df.withColumn(column, (round(col(column)/ col('sum(# CITIES)'), 1)))
      
  return df
```

```
# apply function to temp & precip df
agg_avg_temp_data = get_mean_agg_region(avg_temp_data)
agg_avg_precip_data = get_mean_agg_region(avg_precip_data)
```

```
#change column names
oldColumns = agg_avg_temp_data.schema.names
newColumns_temp = ['CA_ANNUAL_Mean_Temp', 'JAN_Mean_Temp', 'FEB_Mean_Temp', 'MAR_Mean_Temp', 'APR_Mean_Temp', 'MA_Mean_Temp', 'JUN_Mean_Temp', 'JUL_Mean_Temp', 'AUG_Mean_Temp', 'SEP_Mean_Temp', 'OCT_Mean_Temp', 'NOV_Mean_Temp', 'DEC_Mean_Temp', 'Total # CITIES']

agg_avg_temp_data = reduce(lambda agg_avg_temp_data, idx: agg_avg_temp_data.withColumnRenamed(oldColumns[idx], newColumns_temp[idx]), range(len(oldColumns)), agg_avg_temp_data)
agg_avg_temp_data.show()
```

```
#change column names
oldColumns = agg_avg_precip_data.schema.names
newColumns_precip = ['CA_ANNUAL_Mean_Precip', 'JAN_Mean_Precip', 'FEB_Mean_Precip', 'MAR_Mean_Precip', 'APR_Mean_Precip', 'MA_Mean_Precip', 'JUN_Mean_Precip', 'JUL_Mean_Precip', 'AUG_Mean_Precip', 'SEP_Mean_Precip', 'OCT_Mean_Precip', 'NOV_Mean_Precip', 'DEC_Mean_Precip', 'Total # CITIES']

agg_avg_precip_data = reduce(lambda agg_avg_precip_data, idx: agg_avg_precip_data.withColumnRenamed(oldColumns[idx], newColumns_precip[idx]), range(len(oldColumns)), agg_avg_precip_data)
agg_avg_precip_data.show()
```
